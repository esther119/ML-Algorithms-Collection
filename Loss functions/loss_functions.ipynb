```python
def pseudoHuber(error, delta=10):
    return delta**2 * (np.power(1 + (error/delta)**2, 0.5) - 1)
```

```python
def dHuber(error,delta):
    den = (error/delta)**2
    return error / np.power(den+1, 0.5)
```

```python
X = np.linspace(-10,10,10000)
plt.style.use('seaborn-whitegrid')
plt.figure(dpi=150)
delta = 5
plt.xlabel("Error")
plt.ylabel("Huber loss")
plt.plot(pseudoHuber(X,delta), label='Huber')
plt.plot(dHuber(X, delta),label='Derivative')
plt.xticks([0, 2500, 5000, 7500, 10000], [-10, -5, 0, 5,  10])
plt.legend()
plt.savefig("Huber.png")
```

```python
# BCE tests
def BCE(y, y_hat):
    return -(y*np.log(y_hat) + (1-y)*np.log(1-y_hat))

y_hats = [0.8,.6,.95]
y = [1,0,1]
[print(BCE(y[t], y_hats[t])) for t in range(len(y))]
print(sum([BCE(y[t], y_hats[t]) for t in range(len(y))]))
```

```python
# KL Loss
def KL(y, y_hat):
    return y*np.log(y/y_hat)

y = [0,1,0]
y_hat = [0.1, 0.8, 0.1]
res = 0
for i in range(len(y_hat)):
    loss = KL(y[i], y_hat[i])
    res+=loss
    print(loss)
print(res)
```
