```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.stats as stats
```

```python
def normalplot(mean, std, name):
    s = np.random.normal(mean, std, 1000)
    count, bins = np.histogram(s, 30)
    
    plt.plot(bins, 1/(std * np.sqrt(2 * np.pi)) *
                np.exp( - (bins - mean)**2 / (2 * std**2) ),
            linewidth=2, label=name)
```

```python
layers = [64*i for i in range(1,11,1)]
x = np.random.randn(784)
mean = 0
stds = 0

plt.figure(dpi=150)
plt.grid()

for i, n_out in enumerate(layers):
    n_in = x.shape[0]

    std = math.sqrt(1/n_in)
    
    w = np.random.normal(loc=0, scale=std, size=(n_in, n_out))
    x = np.dot(x, w)
    
    normalplot(x.mean(), x.std(), "Layer {}".format(i+1))

    mean += x.mean()
    stds += x.std()

plt.legend(fontsize='xx-small')
print("Mean : {}".format(np.round(mean/len(layers),4)) +  " STD : {}".format(round(stds/len(layers),4)))
```

```python
layers = [64*i for i in range(1,11,1)]
x = np.random.randn(784)
mean = 0
var = 0

plt.figure(dpi=150)
plt.grid()

for i, n_out in enumerate(layers):
    n_in = x.shape[0]

    std = math.sqrt(2/(n_in+n_out))
    
    w = np.random.normal(loc=0, scale=std, size=(n_in, n_out))
    x = np.tanh(np.dot(x, w))
    
    normalplot(x.mean(), x.std(), "Layer {}".format(i+1))

    mean += x.mean()
    var += x.std()**2
    
plt.legend(fontsize='xx-small')
print("Mean : {}".format(np.round(mean/len(layers),4)) +  " Variance : {}".format(round(var/len(layers),4)))
```

```python
layers = [64*i for i in range(1,11,1)]
x = np.random.randn(784)
mean = 0
var = 0

plt.figure(dpi=150)
plt.grid()

for i, n_out in enumerate(layers):
    n_in = x.shape[0]

    std = math.sqrt(2/n_in)
    
    w = np.random.normal(loc=0, scale=std, size=(n_in, n_out))
    x = np.maximum(0,np.dot(x, w))
    
    normalplot(x.mean(), x.std(), "Layer {}".format(i+1))

    mean += x.mean()
    var += x.std()**2

plt.legend(fontsize='xx-small')
print("Mean : {}".format(np.round(mean/len(layers),4)) +  " Variance : {}".format(round(var/len(layers),4)))
```
